```python
x = np.linspace(1,100,100,dtype="int8")
y1 = np.random.randint(1,100,100)
#y2 = np.random.randint(1,100,100)

plt.figure(figsize=(16,4))
plt.plot(x,y1,scalex=True, scaley=True)  
```
