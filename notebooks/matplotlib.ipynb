```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("titanic_dataset.csv")
print(df.tail(10))

```

```python
a = np.random.randint(1,200,30).reshape(10,3)
df = pd.DataFrame(a,columns=["A","B","C"])

x = np.linspace(1,100,100,dtype="int8")
y1 = np.random.randint(1,100,100)
#y2 = np.random.randint(1,100,100)

plt.figure(figsize=(16,4))
plt.plot(x,y1,scalex=True, scaley=True)  
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(x)):
    plt.scatter(x,y1,alpha=.2,color="red",s=.9*2*3.14*y1[idx])
    
```

```python
plt.figure(figsize=(16,4))
plt.plot(range(10),df.A,"-og",label="A") # b,k,m,c,r,g
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(len(x)):
    if(y1[idx]>50):
        plt.scatter(x[idx],y1[idx],alpha=.2,color="red",s=.08*np.pi*y1[idx]**2)
        plt.text(x[idx],y1[idx],y1[idx])

plt.yticks(range(0,351,25))
plt.xticks(range(0,50,2))
plt.xlabel("X-axis",size=20)
plt.ylabel("Y-axis",size=20)
```
