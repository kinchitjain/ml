```python
import pandas as pd
import numpy as np

df=pd.read_csv("C:\\Users\\kinch\\Downloads\\Wholesale_Data.csv")

df.head()
```

```python
df=df.replace(to_replace =["H", "Hot","Hote"],  
                            value ="Hotel")
df=df.replace(to_replace =["C", "Caffe","Caf"],  
                            value ="Cafe")
df=df.replace(to_replace =["R", "Rest"],  
                            value ="Restaurant")
```
